## Project Overview
Predicting if it will rain tomorrow in Australia using weather data from weatherAUS.csv

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import sys

# Add src to path
sys.path.insert(0, str(Path.cwd().parent))

from src.preprocessing import full_pipeline
from src.config import DATA_RAW, MODELS_DIR

# Set style
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (12, 6)

print(f"Data file: {DATA_RAW}")
print(f"Models dir: {MODELS_DIR}")
```

## Load Raw Data

```python
df_raw = pd.read_csv(DATA_RAW)
print(f"Shape: {df_raw.shape}")
print(f"\nFirst rows:")
df_raw.head()
```

```python
# Missing values
missing_pct = (df_raw.isnull().sum() / len(df_raw) * 100).sort_values(ascending=False)
print("Missing values (%):\n", missing_pct[missing_pct > 0])
```

```python
# Target distribution
target_counts = df_raw['RainTomorrow'].value_counts()
print(f"\nTarget distribution:")
print(target_counts)
print(f"\nRain class: {target_counts.get('Yes', 0) / len(df_raw) * 100:.2f}%")
```

## Data Exploration

```python
# Numerical columns statistics
num_cols = df_raw.select_dtypes(include=['number']).columns
df_raw[num_cols].describe()
```

```python
# Categorical columns
cat_cols = df_raw.select_dtypes(include=['object']).columns
for col in cat_cols:
    print(f"\n{col}: {df_raw[col].nunique()} unique values")
    print(df_raw[col].value_counts().head())
```
